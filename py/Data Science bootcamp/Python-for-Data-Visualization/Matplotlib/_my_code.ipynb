```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
## functional method
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r.')
plt.xlabel('Mr X')
plt.ylabel('Mr Y')
```

```python
plt.subplot(1,2,1) # no rows, no columns, plot you are refering to
plt.plot(x,y,'r.')

plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
## OO method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Mr X')
axes.set_ylabel('Mr Y')
axes.set_title('some title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.35])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x,'r.')
axes[1].set_title('Title2')

plt.tight_layout() # put at the end (ALWAYS)
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2,label='x**2',color='red',linewidth=2,linestyle='--',marker='+',markersize=10)
ax.plot(x, x**3,label='x**3')

ax.legend(loc='upper center')

#ax.set_xlim([0,2])
#ax.set_ylim([0,2])
```
