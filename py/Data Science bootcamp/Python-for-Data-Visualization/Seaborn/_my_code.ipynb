```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
# distribution plot
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(data=tips,hue='sex')  ### AWESOME
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# CATEGORICAL PLOTS
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# Matrix (heatmap)
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# correlation
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True)
```

```python
flights.head()
```

```python
flights_matrix = flights.pivot(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(flights_matrix,cmap='coolwarm')
```

```python
sns.clustermap(flights_matrix,cmap='coolwarm')
```
